```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('insurance_pre.csv')
dataset
```

```python
datasets=pd.get_dummies(dataset,drop_first=True)
datasets
```

```python
indep=datasets[['age','bmi', 'children','sex_male', 'smoker_yes']]
dep=datasets[['charges']]
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 1/3, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor( criterion='squared_error',n_estimators=100, max_features= 'log2')
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
